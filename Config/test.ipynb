```python
testD = np.array(testD)
row ,col =testD.shape 
default =set()
data =[]
for i in range(row):
    tmp = testD[i]
    now = []
    for j in range(col):
        a = 'm'+str(j) + tmp[j]
        if tmp[j] != '?':
            default.add(a)
            now.append(str(a))
    data.append(now)
data
```
